```python
import numpy as np
from matplotlib import pyplot as plt
N=500
X = 2*np.random.rand(500,1)
y = 4 +3*X + np.random.randn(500,1)

plt.scatter(X,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```

```python
Theta1 = 3
Theta0 = 4
theta = np.array([[Theta1, Theta0]]).transpose()
print(teata)

oneColumns = np.full((N,1), 1)
X1 = np.column_stack((X, oneColumns))
print(X1[:10, :])
```

```python
predictions = np.matmul(X1, theta)
plt.scatter(X,y)
plt.plot(X, predictions, color="red")
plt.show()
```

```python
def cost(X, y, theta):
    m = X.shape[0]
    predictions = np.matmul(X, theta)
    diff = predictions - y
    cost = (1/2*m)*np.sum(np.square(diff))
    return cost

print(cost(X1, y, theta))
```

```python
def linear_regresstion(X, theta, y, alpha):
    m = X.shape[0]
    num_iterations = 100
    costs = []
    thetas = [theta]
    for i in range(num_iterations):
        predictions = np.matmul(X1, theta)
        errors = predictions - y
        delta = np.matmul(X.T, errors)
        
        theta = theta - (alpha/m)*delta
        cost1 = cost(X,y,theta)
        thetas.append(theta)
        costs.append(cost)
    return costs, thetas
```
