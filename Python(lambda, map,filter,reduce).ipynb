```python
items = [1,2,3,4,5]
squared = list(map(lambda x: x**2, items))
print(squared)
```

```python
names = ['jihoon', 'kang', 'Jihoon', 'Kang']
conterted_tuple = tuple(map(lambda s: str(s).upper(), names))
print(conterted_tuple)

conterted_set = set(map(lambda s: str(s).upper(), names))
print(conterted_set)
```

```python
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]
c = [9, 10, 11, 12]

abc = list(map(lambda x, y, z: x+y+z, a, b, c))
print(abc)
```

```python
funcs = [lambda x: x*x, lambda x: x+x]
for i in range(5):
    value = list(map(lambda x: x(i), funcs))
    print(value)
```

```python
print(range(-5, 5))
print(list(range(-5, 5)))

a = list(filter(lambda x: x<0, range(-5, 5)))
print(a)
```

```python
strObj = 'Hi this is a sample string'
filteredChars = ''.join(filter(lambda x: x not in ['a', 's'], strObj))
print(filteredChars)
#filtered string have to join with str
```

```python
from functools import reduce
#reduce(function, iterable, initializer=None)
reduce((lambda x, y: x*y), [1,2,3])
```
