```python
#2
Korea = pd.Series({'경기수':'14', '금메달':0, '은메달': 1, '동메달':2})
USA = pd.Series({'경기수':'25', '금메달':5, '은메달': 3, '동메달':8})
China = pd.Series({'경기수':'18', '금메달':1, '은메달': 5, '동메달':3})
Japan = pd.Series({'경기수':'15', '금메달':7, '은메달': 1, '동메달':3})
Canada = pd.Series({'경기수':'32', '금메달':8, '은메달': 2, '동메달':6})

group = pd.DataFrame([Korea, USA, China, Japan, Canada], index = ['Korea', 'USA', 'China', 'Japan', 'Canada'])
group
```
